```python
from audioevals.evals import wer_eval, audiobox_eval, vad_eval, pitch_eval, nisqa_eval
from audioevals.utils.audio import AudioData
from pathlib import Path
```

```python
notebook_dir = Path.cwd()          # current working directory
data_dir = notebook_dir.parent / "data"

audio_files = list(data_dir.glob("*.wav"))

data = {}
for audio_file in audio_files:
    print(f"\n\nProcessing {audio_file}")
    audiobox_result = audiobox_eval.run_single_file(str(audio_file))

    pitch_result = pitch_eval.run_single_file(str(audio_file), time_step=0.025)

    nisqa_result = nisqa_eval.run_single_file(str(audio_file))

    data[audio_file.name] = audiobox_result | pitch_result | nisqa_result
 
```

```python
audio_files = sorted(list(data_dir.glob("*.wav")))

print(audio_files)

for t in [0.01, 0.025, 0.05]:
    data = {}
    for audio_file in audio_files:
        pitch_result = pitch_eval.run_single_file(str(audio_file), time_step=t)
        data[audio_file.name] = pitch_result
    for l in [400, 425]:
        max_good = 0
        min_bad = 10000000
        good_lengths = []
        bad_lengths = []
        for key in data:
            if "good" in key:
                max_good = max(max_good, len([x for x in data[key]['pitch_values'] if x >= l]))
                good_lengths.append(len([x for x in data[key]['pitch_values'] if x >= l]))
            else:
                min_bad = min(min_bad, len([x for x in data[key]['pitch_values'] if x >= l]))
                bad_lengths.append(len([x for x in data[key]['pitch_values'] if x >= l]))

        print(t, l, good_lengths, bad_lengths)
```
